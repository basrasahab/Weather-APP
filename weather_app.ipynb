```python
import requests
from gtts import gTTS
from IPython.display import Audio, display

city = input("Enter the city name:\n ")

url = f"https://api.weatherapi.com/v1/current.json?key=14453cd0f27a44ed94b173947251307&q={city}"
response = requests.get(url)
data = response.json()

temperature = data['current']['temp_c']
condition = data['current']['condition']['text']

print(f"Temperature in {city}: {temperature}°C")
print(f"Condition: {condition}")

# Text to speak
weather_report = f"The current temperature in {city} is {temperature} degrees Celsius with {condition}."

# Convert to speech
tts = gTTS(text=weather_report, lang='en')
tts.save("weather.mp3")

# Play the sound in Colab
display(Audio("weather.mp3", autoplay=True))
```
